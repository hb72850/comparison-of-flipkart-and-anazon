```python
#fill na values in Flipkart dataset
flip['retail_price'].fillna(value = 0,inplace = True)
flip['discounted_price'].fillna(value = 0,inplace = True)
```

```python
flip.info()
```

```python
#Changing Product name in upper case in both Flipkart and Amazon dataset to compare with each other
```

```python
flip['product_name'] = flip['product_name'].str.upper()
```

```python
amz['product_name'] = amz['product_name'].str.upper()
```

```python
#Comparing product name 
flip['product_name'] == amz['product_name']
```

```python
#Making a new DataFrame where we take only required clumns from Flipkart
new_flip = flip[['product_name','retail_price','discounted_price']].copy()
```

```python
#Rename the columns So we easily compare with other one
new_flip.rename(columns= {'product_name':'Product name in Flipkart','retail_price':'Retail Price in Flipkart','discounted_price':'Discounted Price in Flipkart'},inplace =True)
```

```python
new_flip.head()
```

```python
#Making a new DataFrame where we take only required clumns from Amazon
new_amz = amz[['product_name','retail_price','discounted_price']].copy()
```

```python
#Rename the columns So we easily compare with other one
new_amz.rename(columns= {'product_name':'Product name in Amazon','retail_price':'Retail Price in Amazon','discounted_price':'Discounted Price in Amazon'},inplace =True)
```

```python
new_amz
```

```python
#using Concat function to joining the to dataset end to end
coparison = pd.concat([new_flip,new_amz],axis =1)
```

```python
coparison
```

```python
#Make a csv file where we easily compare product price and discounted price in both the websites
output = coparison.to_csv(r'E:\hackerearth\int ass\amz and flip\DS - Assignment Part 2 data set\coparison.csv',index =False)
```
